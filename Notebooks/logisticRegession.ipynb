```python
#Pima Indians Diabetes Database
#https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database?resource=download

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.api as sm
```

```python
diabetes_df = pd.read_csv("./input_data/diabetes.csv")
print(diabetes_df.shape)
print(diabetes_df.head())
print(diabetes_df.columns)
print("Total number of cases with diabetes:",np.sum(diabetes_df["Outcome"]))
```

```python
# plt.figure(figsize=(20, 5))
fig, (ax1, ax2) = plt.subplots(1, 2)
# ax1 = plt.axes()
ax1.scatter(diabetes_df["Age"], diabetes_df["Glucose"], color='b', alpha=0.20)
ax1.set_xlabel('Age')
ax1.set_ylabel('Glucose')

ax2.scatter(diabetes_df["BMI"], diabetes_df["Glucose"], color='b', alpha=0.20)
ax2.set_xlabel('BMI')
ax2.set_ylabel('Glucose')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```

```python
smOLS = sm.OLS(diabetes_df["Glucose"], diabetes_df["Age"]).fit()
smOLS.summary()

plt.figure(figsize=(5, 5))
ax = plt.axes()
ax.scatter(diabetes_df["Age"], diabetes_df["Glucose"], color='b', alpha=0.20)
ax.plot(diabetes_df["Age"], smOLS.predict(), color='black', alpha=0.70, linewidth=2)
ax.set_xlabel('Age - 21')
ax.set_ylabel('Glucose')
plt.ylim(top=diabetes_df['Glucose'].max()+10)
plt.show()
```

```python
#y=1+x add_constant add intercept
smOLS = sm.OLS(diabetes_df["Glucose"],sm.add_constant(diabetes_df["Age"])).fit()
smOLS.summary()
```
